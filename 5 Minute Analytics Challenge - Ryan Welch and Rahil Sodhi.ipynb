## c) Versatality

The right-most columns of our dataset show a players overall rating if they were to play in any of the other 10 positions on the field. We created a new column named 'versatility' that takes the average rating of a player if they played anywhere else on the field to represent their effectiveness. 

```python
#With the lack of value that we believe the overall ratings are showing we believe that there is a fatcor missing that isn't being shown
#Below we will create a variable called "versatility" that is an average of the ratings of a player in all the positions they could play in

dfFifa['versatality'] = (dfFifa['ls'] + dfFifa['rs'] + dfFifa['lf'] + dfFifa['rf'] + dfFifa['lam'] + dfFifa['ram'] + dfFifa['lcm'] + dfFifa['rcm'] + dfFifa['lwb'] + dfFifa['cdm'] + dfFifa['rwb'] + dfFifa['lcb'] + dfFifa['rcb'])/13
dfFifa
```

```python
#We graphed only the results for versatility and set the limits to easily see the discrepency in rankings 

avgVersatality = dfFifa.groupby('player_positions')[['versatality']].mean()
avgVersatality.plot.bar(color='darkorange', rot = 3, alpha = 1);
plt.ylim([70,80])
```

```python
#Now in order to run a regression analysis for these attributes we can't have any null values 
#Goalies do not have rankings for these attributes because of the position they play so we will drop all goalies

dfFifa.drop(dfFifa.loc[dfFifa['player_positions']=='GK'].index, inplace=True)
```

```python
#Check that the shape of the data frame has decreased 

dfFifa.info()
```

```python
import sklearn.linear_model as sklm
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
#Now we will create a training set, validation set, and test set in order to conduct regression analysis 

m_reg=sklm.LinearRegression() 

X_m = dfFifa[['pace','passing', 'shooting','defending','dribbling','physic']]
Y_m = dfFifa.overall

X_train_m, X_test_m, Y_train_m, Y_test_m = train_test_split(X_m, Y_m, test_size=0.4, random_state=42)
X_val_m, X_test_m, Y_val_m, Y_test_m = train_test_split(X_test_m, Y_test_m, test_size=0.5, random_state=42)

m_reg.fit(X_train_m, Y_train_m)
```

```python
#We need to check if the data was divided properly

print(X_train_m.shape, Y_train_m.shape)
print(X_val_m.shape, Y_val_m.shape)
print(X_test_m.shape, Y_test_m.shape)
```

```python
#Analyze the coefficients of the data set

print('Intercept:', m_reg.intercept_)

coeff_df = pd.DataFrame(m_reg.coef_, X_m.columns, columns=['Coefficient'])
coeff_df
```

```python
#Now we will analyze how our predictions pair up with the actual values

Y_predict_m = m_reg.predict(X_test_m)

df_m = pd.DataFrame({'Actual': Y_test_m, 'Predicted': Y_predict_m})
df_m
```

```python
#Analyze the error in our predictive modeling

print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test_m, Y_predict_m))
print('Mean Squared Error (MSE):', metrics.mean_squared_error(Y_test_m, Y_predict_m))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test_m, Y_predict_m)))
print("R-Squared:", metrics.r2_score(Y_test_m, Y_predict_m))
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
#Now we use explanatory modeling in order to show all of the important statistical values

formula='overall ~ pace + passing+ shooting+ dribbling + defending+ physic'
model = smf.ols(formula, data=dfFifa) 
tv_reg = model.fit()
tv_reg.summary()
```

## Insights

Passing, dribbling, and physique are the only varaibles that are statistically significant in predicting the variability in the value for overall for the players at a significance level of .05. Defending is not even statistically significant which is the main purpose for the positions we are analyzing and an attribute that the positions we are analyzing rank higher on than most other positions. 13.5% of the variability in overall rating can be predicted by these 6 variables which is not a very high number for the way that tehse variables are utilized in the game itself. Dribbling is statistically significant and has the largest coefficient so this variable takes a large effect on overall rating. The F-statistic is very large so the model we have created is valid. Now we will run a regression with versatility to see how it compares to these variables.
